```python
def lorenz_curve(array):
    array = np.sort(array)  # Sortm valorile n ordine cresctoare
    cumulative_values = np.cumsum(array)  # Suma cumulativ a valorilor
    cumulative_values = np.insert(cumulative_values, 0, 0)  # Adugm 0 la nceput
    cumulative_values = cumulative_values / cumulative_values[-1]  # Normalizm la 1
    gini = 1 - 2 * np.trapezoid(cumulative_values, dx=1/len(array))  # Calculm indicele Gini
    return cumulative_values, gini

cols_to_analyze = ['country', 'region', 'attacktype1', 'weaptype1']  # Numele coloanelor dorite
for col in cols_to_analyze:
    values = df[col].values  # Folosim direct valorile, presupunnd c nu mai sunt lips
    cumulative_values, gini = lorenz_curve(values)
    
    # Crearea graficului pentru curba Lorenz
    plt.figure(figsize=(8, 6))
    plt.plot(np.linspace(0, 1, len(cumulative_values)), cumulative_values, label=f'Curba Lorenz ({col})')
    plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Linia egalitii')
    plt.fill_between(np.linspace(0, 1, len(cumulative_values)), cumulative_values, np.linspace(0, 1, len(cumulative_values)), color='skyblue', alpha=0.4)
    plt.title(f'Curba Lorenz i Indicele Gini pentru {col}\nGini Index: {gini:.4f}')
    plt.xlabel('Proporia populaiei')
    plt.ylabel('Proporia valorilor cumulative')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
from scipy.stats import entropy

def information_quantity(column):
    counts = column.value_counts(normalize=True)
    return entropy(counts, base=2)

entropy_value = information_quantity(df['attacktype1'])
print(f"Entropia pentru 'attacktype1': {entropy_value:.4f}")
entropy_value = information_quantity(df['weaptype1'])
print(f"Entropia pentru 'weaptype1': {entropy_value:.4f}")
entropy_value = information_quantity(df['natlty1'])
print(f"Entropia pentru 'natlty1': {entropy_value:.4f}")
entropy_value = information_quantity(df['country'])
print(f"Entropia pentru 'country': {entropy_value:.4f}")
```

```python
def information_gain(df, feature_col, target_col):
    total_entropy = information_quantity(df[target_col])
    values = df[feature_col].unique()
    
    weighted_entropy = 0
    for value in values:
        subset = df[df[feature_col] == value][target_col]
        weighted_entropy += (len(subset) / len(df)) * information_quantity(subset)
    
    return total_entropy - weighted_entropy

ig_value = information_gain(df, 'attacktype1', 'targtype1')
print(f"Information Gain ntre 'attacktype1' i 'targtype1': {ig_value:.4f}")

ig_value = information_gain(df, 'attacktype1', 'weaptype1')
print(f"Information Gain ntre 'attacktype1' i 'weaptype1': {ig_value:.4f}")

ig_value = information_gain(df, 'natlty1', 'weaptype1')
print(f"Information Gain ntre 'natlty1' i 'weaptype1': {ig_value:.4f}")

ig_value = information_gain(df, 'natlty1', 'country')
print(f"Information Gain ntre 'natlty1' i 'country': {ig_value:.4f}")
df
```

```python
# Model 2  Random Forest + Grid Search
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

# Definim grila de cutare
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2]
}


rf = RandomForestRegressor(random_state=42)

# GridSearch
grid_search_rf = GridSearchCV(estimator=rf, param_grid=param_grid,
                              cv=3, n_jobs=-1, verbose=2, scoring='r2')

grid_search_rf.fit(X_train, y_train)
best_rf_model = grid_search_rf.best_estimator_
y_pred_rf = best_rf_model.predict(X_test)
print("Best parameters found:", grid_search_rf.best_params_)
print("Best R score on validation set:", grid_search_rf.best_score_)
```

```python
# Model 3  XGBoost + Grid Search
from xgboost import XGBRegressor
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform

# Hiperparametrii pentru cutare
param_dist = {
    'n_estimators': randint(100, 300),
    'max_depth': randint(3, 10),
    'learning_rate': uniform(0.01, 0.3),
    'subsample': uniform(0.7, 0.3),
    'colsample_bytree': uniform(0.7, 0.3)
}

# Randomized search
random_search = RandomizedSearchCV(
    estimator=XGBRegressor(random_state=42),
    param_distributions=param_dist,
    n_iter=20,               
    cv=3,
    scoring='neg_mean_squared_error',
    verbose=2,
    n_jobs=-1,
    random_state=42
)


random_search.fit(X_train, y_train)
best_xgb = random_search.best_estimator_
y_pred_xgb = best_xgb.predict(X_test)

print("Cei mai buni hiperparametri pentru XGBoost:")
print(random_search.best_params_)
```

```python
def print_metrics(name, y_true, y_pred):
    print(f"\n--- {name} ---")
    print(f"R: {r2_score(y_true, y_pred):.4f}")
    print(f"RMSE: {np.sqrt(mean_squared_error(y_true, y_pred)):.4f}")
    print(f"MAE: {mean_absolute_error(y_true, y_pred):.4f}")

print_metrics("Linear Regression", y_test, y_pred_lr)
print_metrics("Random Forest", y_test, y_pred_rf)
print_metrics("XGBoost", y_test, y_pred_xgb)
```
